### Faz algumas análises nos valores líquidos dos documentos

```python
df_despesas_2021['valorLiquido'] = df_despesas_2021['valorLiquido'].str.replace(',','.')
df_despesas_2021['valorLiquido'] = df_despesas_2021['valorLiquido'].astype('float64') 
```

```python
# Mostra se há linhas vazias
df_despesas_2021.apply(lambda x: sum(df_despesas_2021['valorLiquido'].isnull()),axis=0)
```

```python
# Separa gastos e média 
gastos = df_despesas_2021['valorLiquido']
media_despesas = lambda gastos: (gastos.mean())
```

```python
# Média
print(media_despesas(gastos))
```

```python
# Mediana
print(gastos.median())
```

```python
# Máximo
print(df_despesas_2021['valorLiquido'].max())
```

```python
# Mínimo
print(df_despesas_2021['valorLiquido'].min())
```

```python
# Máximo menos Mínimo
print(df_despesas_2021['valorLiquido'].max()-df_despesas_2021['valorLiquido'].min())
```

```python
# Estatísticas gerais
print(df_despesas_2021['valorLiquido'].describe())
```

```python
# Total de despesas
df_despesas_2021['valorLiquido'].sum()
```

```python
# Valor que mais se repete
df_despesas_2021['valorLiquido'].mode()
```

```python
# 20 deputados com maiores gastos em cotas
df_despesas_2021.groupby(["nomeParlamentar"]).valorLiquido.sum().reset_index().sort_values("valorLiquido", ascending=False).head(20)
```

```python
# 20 fornecedores com maiores gastos em cotas
df_despesas_2021.groupby(["fornecedor", "cnpjCPF"]).valorLiquido.sum().reset_index().sort_values("valorLiquido", ascending=False).head(20)
```

```python
# 20 tipos de despesas com mais gastos
df_despesas_2021.groupby(["descricao"]).valorLiquido.sum().reset_index().sort_values("valorLiquido", ascending=False).head(20)
```

```python
# Ordenação conjunta - 20 parlamentares + fornecedores com maiores gastos em cotas
df_despesas_2021.groupby(["nomeParlamentar", "fornecedor", "cnpjCPF"]).valorLiquido.sum().reset_index().sort_values("valorLiquido", ascending=False).head(20)
```

```python
# Rotinas para baixar os PDFs dos documentos
```

```python
# Função para criar um nome de arquivo para as notas fiscais, com 2021 + nome do parlamentar + id interno do documento
def limpeza1(linha):
    link = str(linha["urlDocumento"]).strip()
    nome = str(linha["nomeParlamentar"]).strip()
    if link.endswith(".pdf"):
        return "2021"+nome+link[-11:]
    else:
        return ""
```

```python
df_despesas_2021["nome_arquivo"] = df_despesas_2021.apply(limpeza1, axis=1)
```

```python
df_despesas_2021["nome_arquivo"]
```

```python
df_despesas_2021.to_csv('resultados/gastos_cota_parlamentar_2021.csv',index=False)
```
