```python
driver = webdriver.Chrome('D:\Code\chromedriver.exe') 

driver.get(site)
```

```python
# Conta quantas páginas existem 
```

```python
paginas = int(driver.find_element(By.XPATH, "//*[@id='lista_paginate']/ul/li[3]/a/span").text)
print(paginas)
```

```python
#def multiple_replace(string, rep_dict):
#    pattern = re.compile("|".join([re.escape(k) for k in sorted(rep_dict,key=len,reverse=True)]), flags=re.DOTALL)
#    return pattern.sub(lambda x: rep_dict[x.group(0)], string)
```

```python
# Iteração com o número de páginas
for x in range(paginas):

    # Procura os links href
    elems = driver.find_elements_by_xpath("//a[@href]")

    # Itera em cada link encontrado
    for elem in elems:
        #print(elem.get_attribute("href"))
        
        # Testa a captura do link
        # Mostra erro se for um link incorreto no site
        try:
            links = elem.get_attribute("href")
        except:
            print("Erro em: ", elem)
            continue
        
        # Se o link encontrado for para download faz ações
        if "/relatorios/download/" in links:
            link_baixa = links
            print(link_baixa)
            
            # Separa o link por "/" e pega o último valor, número único do relatório
            nome = link_baixa.split("/")[-1]
            
            #titulo = driver.find_element_by_id(nome).text
            #titulo = multiple_replace(titulo, {' ':'_', '-':'_', 'º':'_', '/':'_', ',':'_', '.':'_'}).strip()
            #print(titulo)
            #salva = "D:\\Code\\pinpoint\\repo\\resultados\\cgu\\" + titulo + ".pdf"
            
            # Adiciona final .pdf
            salva = nome + ".pdf"
            
            #conta += 1
            #nome = "cgu__" + str(conta) + ".pdf"
            
            # Faz o download do arquivo e dá o nome com o número único
            wget.download(link_baixa, salva)
        
            #a = input("Espera..........")
    
    # Procura o elemento da próxima página, espera aparecer
    element = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, "//*[@id='lista_paginate']/ul/li[4]/a")))
    # Faz o clique
    element.click()
    # Espera 2 segundos
    time.sleep(2)
```
