```python
import pandas as pd
from datetime import date
from __future__ import unicode_literals
import youtube_dl
import time
import os
```

```python
# Carrega lista criada em pega_lista_videos_yt
```

```python
kwargs = {'sep': ',', 'dtype': str, 'encoding': 'utf-8'}
videos = pd.read_csv("resultados/videos_bolso_22out.csv", **kwargs)
```

```python
# Transforma para formato de data
```

```python
videos['publishedAt'] = pd.to_datetime(videos['publishedAt'])
```

```python
# Cria coluna maiúscula
```

```python
videos['title_copia'] = videos['title']
videos['title_copia'] = videos['title_copia'].str.upper()
```

```python
# Seleciona só vídeos a partir de 2019
```

```python
mask = (videos['publishedAt'].dt.year >= 2019) 
selecionados_atual = videos[mask]
```

```python
selecionados_atual.info()
```

```python
# Seleciona apenas aqueles com "LIVE " no título
```

```python
search_list = ["LIVE "]

mask = selecionados_atual['title_copia'].str.contains('|'.join(search_list))
seleciona = selecionados_atual[mask]
```

```python
# Seta variável com padrão de qualidade do MP3 baixado
# https://stackoverflow.com/questions/49246598/youtube-dl-get-audio-link-with-python/49246653

ydl_opts = {
    'format': 'bestaudio/best',
    'postprocessors': [{
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'mp3',
        'preferredquality': '192',
    }],
}
```

```python
# Itera no dataframe de seleção e baixa cada video_link
# Aponta os vídeos com erro - pode ser erro de conexão ou vídeos banidos ou que só podem ser acessados logado por terem conteúdo ofensivo
# Abaixo tento baixar individualmente os com erro

for num, row in seleciona.iterrows():
    vez = row['video_link']
    print(vez)

    try:
        
        with youtube_dl.YoutubeDL(ydl_opts) as ydl:
            ydl.download([vez])
    
        time.sleep(5)
    except:
        print("////////////////////////////////////////////////////////////////////////////")
        print("ERRO!!!")
        print(vez)
        print("////////////////////////////////////////////////////////////////////////////")
        pass
```
